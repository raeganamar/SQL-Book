# Hitung jumlah ulasan pengguna dan rating rata-rata untuk setiap buku.

```python
query = '''
        select a.title, count(distinct b.review_id) as count, AVG(c.rating) as avg
        from books as a
        inner join reviews as b on a.book_id = b.book_id
        inner join ratings as c on a.book_id = c.book_id
        group by 1
        order by count desc, avg desc
        '''
result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

# Identifikasi penerbit yang telah menerbitkan jumlah buku terbanyak, dengan jumlah halaman lebih dari 50

```python
query = '''
        SELECT publisher, COUNT(book_id)
        FROM books
        JOIN publishers
        ON books.publisher_id = publishers.publisher_id
        WHERE books.num_pages > 50
        GROUP BY publisher
        ORDER BY COUNT(book_id) DESC
        LIMIT 1
        '''

result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

Kesimpulan sementara: 
- Penerbit yang telah menerbitkan jumlah buku terbanyak dengan syarat jumlah halaman > 50 adalah Penguin Books sebanyak 42 buku

# Identifikasi penulis dengan rating rata-rata buku tertinggi: temukan buku dengan rating minimal 50


```python
query = '''
        SELECT b.author,AVG(c.rating) as avg, COUNT(c.rating) as count
        from books as a
        join authors as b on a.author_id = b.author_id
        join ratings as c on a.book_id = c.book_id
        group by 1
        having count(c.rating) > 50
        order by avg desc
        limit 1
        '''

result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

Penulis dengan rating rata-rata buku tertinggi dengan rating minimal 50 adalah `J.K. Rowling/Mary GrandPré`

# Hitung jumlah rata-rata teks ulasan di antara pengguna yang memberi rating terhadap lebih dari 50 buku.

```python
query = ''' 
        SELECT 
             AVG(subquery.review_count) AS avg_text_reviews
        FROM 
            (select reviews.username,count(reviews.review_id) as review_count
            from 
            reviews
             WHERE reviews.username IN (SELECT 
                                    ratings.username
                                     FROM 
                                    ratings 
                                     GROUP BY 
                                    ratings.username 
                                     HAVING COUNT(ratings.rating_id) > 50)
             group by 1) as  subquery
        '''

result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

jumlah rata-rata teks ulasan di antara pengguna yang memberi rating terhadap lebih dari 50 buku adalah 24.33
